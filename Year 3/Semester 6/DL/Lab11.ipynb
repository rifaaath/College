```python
#Regularization techniques
```

```python
from sklearn.datasets import make_moons

x, y = make_moons(100, noise=.25)
```

```python
import matplotlib.pyplot as plt

plt.scatter(x[ : , 0],x[ : , 1], c=y)
plt.show()
```

```python
from keras.models import Sequential
from keras.layers import Dense
from keras.regularizers import l1

model1 = Sequential()
model1.add(Dense(128, input_dim = 2, activation = 'relu'))
model1.add(Dense(128, activation = 'relu'))
model1.add(Dense(1, activation = 'sigmoid'))

model2 = Sequential()
model2.add(Dense(128, input_dim = 2, activation = 'relu', kernel_regularizer = l1(.001)))
model2.add(Dense(128, activation = 'relu', kernel_regularizer = l1(.001)))
model2.add(Dense(1, activation = 'sigmoid'))

model1.summary()
model2.summary()
```

```python
model1.compile(optimizer = 'adam', loss = 'binary_crossentropy', metrics = ['accuracy'])
model2.compile(optimizer = 'adam', loss = 'binary_crossentropy', metrics = ['accuracy'])
hist1 = model1.fit(x, y, epochs=100, validation_split=0.2)
hist2 = model2.fit(x, y, epochs=100, validation_split=0.2)
```

```python
from mlxtend.plotting import plot_decision_regions

plot_decision_regions(x, y, clf=model1)
plt.legend()
plt.show()

plot_decision_regions(x, y, clf=model2)
plt.legend()
plt.show()
```

```python
#Plotting
plt.plot(hist1.history['accuracy'], label='Model 1 Train')
plt.plot(hist1.history['val_accuracy'], label='Model 1 Test')
plt.plot(hist2.history['accuracy'], label='Model 2 Train')
plt.plot(hist2.history['val_accuracy'], label='Model 2 Test')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()
plt.show()
```

```python
#Plotting
plt.plot(hist1.history['loss'], label='Model 1 Train')
plt.plot(hist1.history['val_loss'], label='Model 1 Test')
plt.plot(hist2.history['loss'], label='Model 2 Train')
plt.plot(hist2.history['val_loss'], label='Model 2 Test')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()
plt.show()
```

```python
print('Accuracy for model 1:', hist1.history['accuracy'][-1])
print('Accuracy for model 2:', hist2.history['accuracy'][-1])
```
