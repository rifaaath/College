10. Fetch the real time top rated movie reviews from tmdb site and demonstrate how to save the data as .csv format  

```python
import pandas as pd
import requests
```

```python
response = requests.get('https://api.themoviedb.org/3/movie/top_rated?api_key=7bec7dc120b64b787d6b6ec386fca08a&language=en-US&page=1')
response.json()
```

```python
df = pd.DataFrame()
df
```

```python
for i in range(1,3):
    response=requests.get('https://api.themoviedb.org/3/movie/top_rated?api_key=7bec7dc120b64b787d6b6ec386fca08a&language=en-US&page={}'.format(i))
    temp_df=pd.DataFrame(response.json()['results'])[['id','title','overview', 'release_date','popularity','vote_average','vote_count']]
    df=df.append(temp_df, ignore_index=True)
print(df.shape)
# To convert into csv
df.to_csv('movies.csv')
```
